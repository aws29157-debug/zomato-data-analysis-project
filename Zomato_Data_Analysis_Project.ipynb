### Merge Both Datasets

## EDA By Graphs

### Visualizing Restaurant Distribution by Country

```python
country_counts = final_df['Country'].value_counts()

plt.figure(figsize=(10, 6))
sns.barplot(x=country_counts.index, y=country_counts.values)

plt.xticks(rotation=90)
plt.title('Number of Restaurants by Country')
plt.xlabel('Country')
plt.ylabel("Number of Restaurants")

plt.show()
```

### Rating Distribution Analysis

```python
# Distribution of Aggregate Rating

final_df["Aggregate rating"].describe()
```

### Rating Distribution Graph

```python
# Histogram of Restaurant Ratings
plt.figure(figsize=(8, 6))

sns.histplot(final_df['Aggregate rating'], bins=30, kde=True)

plt.title('Distribution of Restaurant Ratings')
plt.xlabel('Rating')
plt.ylabel("Number of Restaurants")

plt.show()
```

### Price Range vs Rating

```python
# Price Range vs Restaurant Ratings

plt.figure(figsize=(10, 6))

sns.boxplot(x="Price range",
            y="Aggregate rating",
            data=final_df)

plt.title("Price Range vs Restaurant Ratings")

plt.show()
```

### Delivery vs Rating Analysis

```python
# Online Delivery vs Restaurant Rating
plt.figure(figsize=(8, 6))

sns.boxplot(x="Has Online delivery",
            y="Aggregate rating",
            data=final_df)

plt.title("Online Delivery vs Restaurant Rating")

plt.show()
```

# Restaurant Success Analysis
## Votes vs Restaurant Rating

#### Observation: Restaurants with higher customer votes tend to have higher ratings.

### Business Meaning: Customer engagement plays an important role in restaurant reputation.

```python
# Customer Engagement vs Restaurant Rating
plt.figure(figsize=(8, 6))

sns.scatterplot(
    x="Votes",
    y="Aggregate rating",
    data=final_df
)

plt.title("Votes vs Restaurant Rating")

plt.show()
```

### Cuisine vs Rating

```python
top_cuisines = final_df['Cuisines'].value_counts().head(10).index

plt.figure(figsize=(10, 6))

sns.boxplot(
    x="Aggregate rating",
    y="Cuisines",
    data=final_df[final_df['Cuisines'].isin(top_cuisines)]
)

plt.title("Top Cuisines vs Ratings")

plt.show()

```

### Distribution of Customer Engagement (Votes)

```python
plt.figure(figsize=(8, 6))

sns.histplot(
    final_df['Votes'], bins=50)

plt.title("Distribution of Votes")

plt.show()
```

### 2. Delivery services should be expanded to increase restaurant visibility.

### 3. New restaurants should focus on popular cuisines such as North Indian and Chinese.

### 4. Restaurants in smaller cities may have growth opportunities as the market is concentrated in metro cities.

## Conclusion

###This analysis shows that restaurant popularity is strongly related to customer engagement and consistent service quality.
